```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# import PIL as im

# Imports PIL module
from PIL import Image

# open method used to open different extension image file
# im = Image.open("fig.png")
img = cv2.imread('fig.png')

negative_img = 255-img

# This method will show image in any image viewer
cv2_imshow( negative_img)
# im.show()
```

```python
flip_img = cv2.flip(img, 1)
cv2_imshow(flip_img)
```

```python
#converting to gray_scale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
min_intensity, max_intensity = np.min(gray_img), np.max(gray_img)

contrast_stretched_img = 255 * (gray_img - min_intensity) / (max_intensity - min_intensity)

cv2_imshow(contrast_stretched_img.astype(np.uint8))
```

```python
cv2.imread('fig.png', cv2.IMREAD_GRAYSCALE)

threshold_value = 200

_, thresholded_img = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

cv2_imshow(thresholded_img)
```
